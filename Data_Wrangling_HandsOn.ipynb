# Data Wrangling

**Data Wrangling** also called **data cleaning**, **data remediation** or **data muging**- refers to a  variety of process designed to transform raw data into mode readlily used format. The process includes connecting to data sources, reformatting the information so it's consistent, removing duplicates, merging disparate sources and filtering out unneeded "noise" in large dataset.

## Combining Data



### Concat & append

The difference between concat and append is that **concat can combine dataframes by row or column**, whereas append can only combine rows

Computationally insignificant difference: https://stackoverflow.com/questions/15819050/pandas-dataframe-concat-vs-append

```python
#membuat dataframe

# Membuat dataframe pertama dari struktur data dictionary
df1 = pd.DataFrame({"a":[1, 2, 3, 4],
                    "b":[5, 6, 7, 8]})
  
# Membuat dataframe kedua dengan tambahan kolom c dari struktur data dictionary
df2 = pd.DataFrame({"a":[5,6,7],
                    "b":[9, 10, 11]})
```

```python
df1
```

```python
df2
```

```python
#concat menggabungkan baris tanpa reset index
df3 = pd.concat([df1,df2])
df3
```

```python
#concat menggabungkan baris
df3 = pd.concat([df1,df2]).reset_index(drop=True)
df3
```

```python
#concat menggabungkan columns
df3 = pd.concat([df1,df2], axis=1).reset_index(drop=True)
df3
```

```python
#append
df3 = df1.append(df2, ignore_index=True)
df3
```

### Merge Data

```python
# Membuat tabel transaksi dari struktur data dictionary
transaksi = pd.DataFrame({"user_id":[1, 2, 3, 4],
                     "product_id":[5,2,2,1],
                     "user_name":["Anto", "Bimo", "Cynthia", "Dimas"]})

# Membuat tabel kota tinggal dari struktur data dictionary
kota = pd.DataFrame({"user_id":[1, 2, 3, 8, 9],
                     "city":["Jakarta", "Medan", "Bandung", "Depok", "Surabaya"]})

# Membuat tabel product dari struktur data dictionary
product = pd.DataFrame({"product_id":[1, 2, 3, 4,5],
                       "product":["Piring", "Gelas", "Sendok", "Garpu","Mangkok"]})
```

```python
transaksi
```

```python
kota
```

```python
product
```

```python
#merge pertama antara table transaksi dan kota untuk mengetaui asal customer. merge dengan menggunakan left join
merge_1=transaksi.merge(kota, how="left", on=['user_id'])
merge_1
```

```python
transaksi.merge(kota, how="right", on=['user_id'])
```

```python
transaksi.merge(kota, how="inner", on=['user_id'])
```

```python
transaksi.merge(kota, how="outer", on=['user_id'])
```

```python
#merge_final adalah gabungan antara table merge_1 dengan product untuk mengetahui product yang di beli
merge_final=merge_1.merge(product,how='left',on=['product_id'])
merge_final
```

## Modify/ Aggregation Data

### Group By

```python
#aggregate based on 1 column 

#menghitung jumlah customer berdasarkan eduaction
jum=df.groupby(['Education'])[['Income']].mean()
jum
```

```python
#menghitung jumlah customer berdasarkan eduaction
jum=df.groupby(['Education'])[['Marital_Status']].count()
jum
```

```python
#menghitung jumlah unique customer berdasarkan education

df.groupby(['Education'])[['ID']].nunique()
```

```python
#aggregate multiple columns

#menghitung jumlah customer berdasarkan education dan marital status
df.groupby(['Education','Marital_Status'])[['Income']].mean()
```

```python
#aggregate using nested columns 

#menampilkan lebih banyak aggregasi dengan nested columns
df.groupby(['Education']).agg({
'Income':['mean','min','max'],'ID':'count'})
```

### Pivot tables

pivot table dapat melakukan aggregasi terhadap 2 variable katagori

```python
#single index (default mean)
pd.pivot_table(data=df,index=['Education'])
```

```python
#multiple index
pd.pivot_table(df,index=['Education','Marital_Status'])
```

```python
#different aggregate functions
pd.pivot_table(df,index=['Education','Marital_Status'],aggfunc={'Recency':np.mean,'Income':np.sum})
```

```python
#Find the relationship between features with columns parameter
table = pd.pivot_table(df,index=['Response'],columns=['Education'],values=['ID'],aggfunc='count')
table
```
